Read and clean the data

```r
getdata.moth <- function(input)
{
    # some of this code is taken from Michael T's functions
    
    # moth data
    mothdata <- tibble::as_tibble(read.csv("/data/ecn/ECN_IM1.csv"))
    
    # process
    dd = mothdata %>%
    # change year format yy to yyyy
    dplyr::mutate(YEAR=as.integer(str_sub(SDATE, start= -2))) %>%
    dplyr::mutate(YEAR=if_else( YEAR <= 50,YEAR+2000,YEAR+1900)) %>%
    dplyr::mutate(DATE= as.Date(SDATE, format= "%d-%b-%y")) %>%
    # drop YEAR and SDATE columns
    dplyr::select(-YEAR,-SDATE) %>%
    # filter by (selected) sites
    dplyr::filter(SITECODE %in% input$sites) %>%
    # get dates in specified date range
    dplyr::filter(
        as.Date(DATE) >= as.Date(input$daterange[1]) &
        as.Date(DATE) <= as.Date(input$daterange[2]) 
    ) %>%
    # remove rows with "Q" in FIELDNAME
    dplyr::filter(!str_detect(FIELDNAME, "Q"))
    
    return(dd)
}

data.moth <- getdata.moth(input)

print(data.moth, n=5)
```

# Daily counts of all moths
- The data for the following traps, T02-1, T04-1, T04-3, T10-1, T10-2 and T12-1, is sparse. We will proceed without these traps.
- Hence, we proceed with the following traps, "T01-1", "T02-2", "T03-1", "T04-2", "T05-1", "T06-1", "T06-2", "T07-1", "T08-1", "T09-1", "T10-3".

```r
getdata.moth.dailycount <- function(input)
{
    # read and clean moth data
    data.moth <- getdata.moth(input)
    
    return(
        data.moth %>%
        # add TRAPCODE column
        mutate(
            TRAPCODE = paste0(SITECODE, "-", LCODE)
        ) %>%
        # Aggregate over FIELDNAME (count all moths)
        group_by(TRAPCODE, DATE) %>%
        summarize(VALUE = sum(VALUE)) %>%
        ungroup()
    )
}

options(repr.plot.width = 24, repr.plot.height = 12)
data.moth.dailycount <- getdata.moth.dailycount(input)
plot.data.moth.dailycount <- data.moth.dailycount %>%
    ggplot(aes(x = DATE, y = VALUE)) + 
    geom_point(size = 0.3) + 
    facet_wrap(
        TRAPCODE~.,
        scales = "free_y"
    ) + 
    mytheme
plot.data.moth.dailycount
```

# Interval between traps emptied
  - There is some change in emptying freqeuency. Around 2004 the interval between traps being emptied appears to change, the time of change is highlighted by the solid (blue) vertical line
  - There is some disruption to emptying traps begining around the start of 2000. This may be related to the foot-and-mouth outbreak.
  - There are very large sampling intervals

```r
getdata.mothtrap.dailycount <- function(input)
{
    sites <- unique(do.call(c, lapply(input$traps, function(x){ strsplit(x, "-")[[1]][1]})))
    
    # read and clean moth data
    data.moth <- getdata.moth(
        list(
            daterange = input$daterange,#
            sites = sites
        )
    )
    
    return(
        data.moth %>%
        # add TRAPCODE column
        mutate(
            TRAPCODE = paste0(SITECODE, "-", LCODE)
        ) %>%
        filter(
            TRAPCODE %in% input$traps
        ) %>%
        # Aggregate over FIELDNAME (count all moths)
        group_by(TRAPCODE, DATE) %>%
        summarize(VALUE = sum(VALUE)) %>%
        ungroup()
    )
}
traps.selected <- c("T01-1", "T02-2", "T03-1", "T04-2", "T05-1", "T06-1", "T06-2", "T07-1", "T08-1", "T09-1", "T10-3")

data.mothtrap.dailycount <- getdata.mothtrap.dailycount(
    list(
        daterange = daterange,
        traps = traps.selected
    )
)

# add interval between observations
data.mothtrap.dailycount <- data.mothtrap.dailycount %>%
    # calculate the interval between observations 
    # note this accounts for empty traps as FIELDNAME XX included
    group_by(TRAPCODE) %>%
    arrange(DATE) %>%
    mutate(
        interval = c(NA, diff(DATE))
    ) %>%
    ungroup() %>%
    filter(!is.na(interval))

options(repr.plot.width = 24, repr.plot.height = 14)
plot.data.mothtrap.interval <-  data.mothtrap.dailycount %>%
    # plot
    ggplot(aes(x = DATE, y = interval)) + 
    geom_point(size = 0.3) +
    geom_vline(
        data = data.frame(loc = as.Date(c("2004-04-01"))),
        aes(xintercept = loc),
        colour = "blue"
    ) + 
    geom_vline(
        data = data.frame(loc = as.Date(c("2000-01-01"))),
        aes(xintercept = loc),
        linetype = 2,
        colour = "red"
    ) + 
    scale_color_gradientn(colours = rainbow(3)) +
    facet_grid(
        #TRAPCODE ~ .,
        row = vars(TRAPCODE),
        #col = vars(VARIABLE),
        scales = "free_y"
    ) + 
    mytheme
plot.data.mothtrap.interval
```

# Interval range for each trap
 - The largest interval between empyting traps exceeds 42 days for all traps
 - It is reasonable to adjust the counts by the number of days that pass between emptying. But, including and adjusting data for very large intervals may be misleading.
 - Determine a maximum interval threshold

```r
data.mothtrap.dailycount %>%
    group_by(TRAPCODE) %>%
    summarize(
        interval.min = min(interval),
        interval.max = max(interval)
    )
```

# Interval threshold
 - Having a maximum interval of 5 days removes no more than 9% of observations for any trap
 - 8 out of 11 traps lose less than 4% of data

```r
threshold = 5
gettbl.threshold <- function(tbl, threshold)
{
    return(
        tbl %>%
        group_by(TRAPCODE) %>%
        summarize(
            nobs = n(),
            interval.min = min(interval),
            interval.max = max(interval),
            nlessthresh = sum(interval <= threshold),
            percent.kept = 100*nlessthresh/nobs,
            percent.removed = 100*(nobs-nlessthresh)/nobs
        ) %>%
        mutate(interval.threshold = threshold)
    )    
}

gettbl.threshold(data.mothtrap.dailycount, 5)

data.mothtrap.dailycount.threshold <- data.mothtrap.dailycount %>%
    filter(interval < threshold) 

plot.data.mothtrap.dailycount.threshold <- data.mothtrap.dailycount.threshold %>%
    ggplot(aes(x = DATE, y = VALUE)) + 
    geom_point(size = 0.3) + 
    facet_wrap(
        TRAPCODE~.,
        scales = "free_y"
    ) + 
    mytheme
plot.data.mothtrap.dailycount.threshold
```

# Range of counts affected
 - The range of maximum counts changes significantly for the following traps T01-1 (305 --> 240) and T10-3 (811 --> 381).
 - The range of maximum counts changes for the reminaing traps as follows, T07-1 (320 --> 302) and T09-1 (128 --> 125).
 - The remaining trap maximum counts does not change

```r
data.mothtrap.dailycount.threshold %>%
    group_by(TRAPCODE) %>%
    summarize(
        thresh.min = min(VALUE),
        thresh.max = max(VALUE)
    ) %>%
    left_join(
        data.mothtrap.dailycount %>%
        group_by(TRAPCODE) %>%
        summarize(
            min = min(VALUE),
            max = max(VALUE)
        ),
    by = "TRAPCODE")
```

# Daily moth rate

```r
getdata.mothtrap.dailyratethresh <- function(input)
{
    data.mothtrap.dailycount <- getdata.mothtrap.dailycount(input)
    
    data.mothtrap.dailycount <- data.mothtrap.dailycount %>%
        # calculate interval
        group_by(TRAPCODE) %>%
        arrange(DATE) %>%
        mutate(
            interval = c(NA, diff(DATE))
        ) %>%
        # remove rows with NA interval value
        filter( !is.na(interval) ) %>%
        ungroup()
    
    # print inpact of thresholding interval
    message("Table below summarises impact of thresholding")
    print(gettbl.threshold(data.mothtrap.dailycount, input$interval.thresh))
    
    # remove values with interval above threshold and calculate rate
    newdata <- data.mothtrap.dailycount %>%
        # remove datapoints with large interval
        filter( interval <= input$interval.thresh) %>%
        # calculate rate
        mutate(VALUE = VALUE/interval)
    
    message("Table below shows how the minimum and maximum counts change after thresholding interval")
    print(
        newdata %>%
        group_by(TRAPCODE) %>%
        summarize(
            newvalue.min = min(VALUE*interval),
            newvalue.max = max(VALUE*interval)
        ) %>%
        left_join(
            data.mothtrap.dailycount %>%
            group_by(TRAPCODE) %>%
            summarize(
                value.min = min(VALUE),
                value.max = max(VALUE)
            ),
        by = "TRAPCODE"
        )
    )
    
    return(newdata)
}

data.mothtrap.dailyratethresh <- getdata.mothtrap.dailyratethresh(
    list(
        traps = traps.selected,
        daterange = daterange,
        interval.thresh = 5
    )
)
plot.data.mothtrap.dailyratethresh <- data.mothtrap.dailyratethresh %>%
    ggplot(aes(x = DATE, y = VALUE, colour=interval>1)) + 
    geom_point(size = 0.3) +
    facet_wrap(
        TRAPCODE ~ .,
        scales = "free_y"
    ) + mytheme
```

# Monthly moth rate

```r
options(repr.plot.width = 24, repr.plot.height = 16)

getdata.mothtrap.monthlyratethresh <- function(input)
{
    data.mothtrap.dailyratethresh <- getdata.mothtrap.dailyratethresh(input)
    
    # calculate average monthly rate
    return(
        data.mothtrap.dailyratethresh %>%
        mutate(
            year = lubridate::year(DATE),
            monthname = lubridate::month(DATE, label=TRUE, abbr=FALSE)
        ) %>%
        group_by(TRAPCODE, year, monthname) %>%
        summarize(
            VALUE = mean(VALUE)
        ) %>%
        mutate(
            DATE = as.Date(paste(year,monthname,1,sep='-'), format="%Y-%B-%d")
        ) %>%
        ungroup() %>%
        select(DATE, VALUE, TRAPCODE)
    )
}

data.mothtrap.monthlyratethresh <- getdata.mothtrap.monthlyratethresh(
    list(
        traps = traps.selected,
        daterange = daterange,
        interval.thresh = 5
    )
)

seasons <- data.mothtrap.monthlyratethresh %>%
    mutate(
        month = lubridate::month(DATE)
    ) %>%
    group_by(TRAPCODE, month) %>%
    summarize(avg = mean(VALUE)) %>%
    ungroup()

tmp <- data.mothtrap.monthlyratethresh %>%
    mutate(month = lubridate::month(DATE)) %>%
    left_join(
        seasons,
        by = c("TRAPCODE", "month")
    ) %>%
    mutate(
        deseason = VALUE - avg
    ) %>%
    group_by(TRAPCODE, month) %>%
    mutate(
        ssd = sd(deseason)
    ) %>%
    ungroup() %>%
    mutate(std = deseason/ssd)

tmp %>%
    #ggplot(aes(x = DATE, y = deseason)) +#, colour = month)) + 
    ggplot(aes(x = DATE, y = std)) +#, colour = month)) + 
    geom_point(size = 1) + 
    geom_line() +
    geom_hline(yintercept=0) +
    scale_color_gradientn(colours = rainbow(3)) +
    #eom_hline(data = tmp, aes(yintercept = avg)) +
    facet_grid( 
        rows = vars(TRAPCODE),#TRAPCODE ~ .,
        cols = vars(month),
        scales = "free_y"
    )

tmp %>%
    filter( !is.na(std) ) %>%
    #ggplot(aes(x = DATE, y = deseason)) +#, colour = month)) + 
    ggplot(aes(x = DATE, y = std)) +#, colour = month)) + 
    geom_point(size = 1) + 
    geom_line() +
    geom_hline(yintercept=0) +
    scale_color_gradientn(colours = rainbow(3)) +
    #eom_hline(data = tmp, aes(yintercept = avg)) +
    facet_grid( 
        rows = vars(TRAPCODE),#TRAPCODE ~ .,
        scales = "free_y"
    )
```

```r
options(repr.plot.width = 24, repr.plot.height = 12)

data <- tmp %>% filter( !is.na(std) )
library(changepoint.msr)
keys <- sort(unique(data$TRAPCODE))
keys <- c("T01-1", "T03-1", "T04-2", "T08-1")#, "T05-1", "T06-2", "T09-1")
cpt.out <- cpt.msr(
    data = data,
    test.stat = rep("normal_meanvar", length(keys)),
    penalty = rep("bic", length(keys)),
    #key_value = "deseason",
    key_value = "std",
    key_index = "DATE",
    key_column = "TRAPCODE",
    keys = keys
)

cpt.out@data %>%
    ggplot(aes(x=index, y=value)) +
    geom_point(aes(colour=lubridate::month(index))) + 
    scale_color_gradientn(colours = rainbow(3)) +
    geom_line() +
    facet_grid(row = vars(id)) +
    geom_vline(
        data = cpt.out@msr.cpts,
        aes(xintercept = loc)
    )
```
