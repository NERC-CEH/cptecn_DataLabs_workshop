# Average monthly weather variable

```r
tmp <- inputDataMA %>% 
    dplyr::filter(SITECODE=="T06", FIELDNAME=="DRYTMP") %>%
    tidyr::pivot_wider(names_from = FIELDNAME, values_from = daily) %>%
    dplyr::mutate( yearmonth= tsibble::yearmonth(DATE) ) %>%
    dplyr::group_by( yearmonth ) %>%
    dplyr::summarize( monthly_avg = mean(DRYTMP, na.rm=TRUE) ) %>%
    dplyr::ungroup()

print( "tmp" )
print( tmp )
```

# Ensure all months are present

```r
yearmonth_min <- min( tmp$yearmonth )
yearmonth_max <- max( tmp$yearmonth )

allData <- tibble::tibble(yearmonth = seq(yearmonth_min, yearmonth_max, by = 1)) %>%
    dplyr::left_join(tmp) %>%
    dplyr::mutate(DATETIME = lubridate::as_date(format(yearmonth, "%Y-%m-%d"))) %>%
    dplyr::mutate(month = lubridate::month(DATETIME)) 

print( "allData" )
print( allData )
```

Fill missing values with month median

```r
missing <- sum( is.na(allData$monthly_avg) )
if( missing > 0 )
{
    print( c("total missing values:", missing) )
    
    monthly_median <- allData %>%
        dplyr::group_by(month) %>%
        dplyr::summarize( median_month = median(monthly_avg, na.rm=TRUE) ) %>%
        dplyr::ungroup() %>%
        dplyr::arrange(median_month)
    monthly_median
    
    allData <- monthly_median %>% 
        dplyr::left_join(allData) %>% 
        dplyr::mutate(missing = case_when(is.na(monthly_avg) ~ TRUE,
                                          !is.na(monthly_avg) ~ FALSE),
                      monthly_avg = case_when(is.na(monthly_avg) ~ median_month,
                                              !is.na(monthly_avg) ~ monthly_avg))
} else
{
    allData <- allData %>% dplyr::mutate(missing=FALSE)
}
```

# Plot the data

```r
allData %>% ggplot2::ggplot(aes(x = DATETIME, y = monthly_avg, color=missing) ) +
    geom_point() +
    scale_color_manual(values = c("#1f77b4", "#d62728"))
```
